```python
import matplotlib.pyplot as plt

# Step 1
# Data samples
C1 = [(2, 3), (3, 3), (3, 4), (1, 4), (4, 1), (4, 3)]
C2 = [(0, 0), (0, 3), (1, 1), (1, 2), (2, 1), (2, 2)]


# Create empty x and y lists
x1 = []
y1 = []
x2 = []
y2 = []


# Split x and y for Class C1
for point in C1:
    x1.append(point[0])
    y1.append(point[1])


# Split x and y for Class C2
for point in C2:
    x2.append(point[0])
    y2.append(point[1])


# Steps 2 and 3
# Asking user for input on thresholds and preventing common errors such as entering characters instead of numbers
def get_float(prompt):
    while True:
        # Step 7: allowing user to exit by typing 'x'
        value = input(prompt)
        if value.lower() == 'x':
            return 'exit'
        try:
            num = float(value)
            return num
        except ValueError:
            print("That is not a float -- please try again")


# Step 4
# Apply classification rule to each point
def classify_points(data, thx, thy):
    predicted = []
    for point in data:
        if point[0] >= thx and point[1] >= thy:
            predicted.append('C1')
        else:
            predicted.append('C2')
    return predicted


# Step 4 (continued)
# Calculate accuracy
def calculate_accuracy(true_labels, predicted_labels):
    correct = 0
    for i in range(len(true_labels)):
        if predicted_labels[i] == true_labels[i]:
            correct += 1
    return correct / len(true_labels)
```

```python
# Step 5
# Plotting results with threshold lines and fixed axis range
def plot_results(x1, y1, x2, y2, thx, thy):
    # Plotting Class C1 with red circles, Class C2 with pink squares
    plt.plot(x1, y1, 'ro', label='Class C1')
    plt.plot(x2, y2, 'ms', label='Class C2')


    # Plotting threshold lines
    # vertical line
    plt.plot([thx, thx], [-1, 5], 'k--', label='thx')
    # horizontal line
    plt.plot([-1, 5], [thy, thy], 'g--', label='thy')


    # Adding labels and legend (as shown in lecture)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Data Samples: C1 vs C2')
    plt.legend()
    plt.grid(True)


    # Setting axis range (as recommended in tip)
    plt.xlim(-1, 5)
    plt.ylim(-1, 5)


    # Showing the plot
    plt.show()



# Step 6 and Step 7
# Creating a loop to continuously repeat steps 4 and 5 and enter different sets of thresholds each time
while True:
    thx = get_float("Enter threshold for x (thx) or 'x' to exit: ")
    if thx == 'exit':
        print("You exited by typing 'x' for thx.")
        break

    thy = get_float("Enter threshold for y (thy) or 'x' to exit: ")
    if thy == 'exit':
        print("You exited by typing 'x' for thy.")
        break


    # Step 4
    data = C1 + C2
    true_labels = ['C1'] * len(C1) + ['C2'] * len(C2)
    predicted_labels = classify_points(data, thx, thy)
    accuracy = calculate_accuracy(true_labels, predicted_labels)


    # Printing the result
    print("Classification accuracy is:", accuracy)


    # Step 5
    plot_results(x1, y1, x2, y2, thx, thy)
```

Step 8:

Based on my observations from the results above, the highest classification accuracy was 0.9167 or 91.67%, using a threshold of x = 1 and y = 3.
This threshold combination gave the most accurate classification I could achieve from the 12 data points.
